```python
import pandas as pd
cases_df = pd.read_json("https://zissou.infosci.cornell.edu/convokit/datasets/supreme-corpus/cases.jsonl", lines=True)
cases_df.head()
```

**Separate components of corpus:**
- Utterances
- Converstions
- Speakers

```python
#UTTERANCES
utterances_df = corpus.get_utterances_dataframe()
utterances_df.head()
```

```python
#CONVERSATIONS
conversations_df = corpus.get_conversations_dataframe()
conversations_df.head()
```

```python
#SPEAKERS
speakers_df = corpus.get_speakers_dataframe()
speakers_df.head()
```

**Analysis:**
Winning vs Losing side features:
- Sentiment Score
- Lawyer Response Length to Justice's Questions
- Citing precedent

```python
#winning side distribution
print(utterances_df["meta.side"].value_counts())
print(utterances_df["meta.speaker_type"].value_counts())
```

```python
conversations_df["meta.win_side"].value_counts()
```

*Interpretation*

---



1= how many cases were won by the petitioner (4930)

0= how many cases were won by the respondent (2868)

-1= how many cases have no winner information (2)


```python
#get all the advocate utterances and save them as their own separate dataframe
advocate_utterances = utterances_df[utterances_df["meta.speaker_type"] == "A"].copy()

#look up winning side for each utterance based on which conversation it came from
advocate_utterances = advocate_utterances.merge(
    conversations_df[["meta.win_side"]],
    left_on="conversation_id",
    right_index=True,
    how="left"
)

advocate_utterances.head()
```

*Interpretation*


---


advocate_utterances:
has all the advocate speech, but doesn't know who won


conversations_df[["meta.win_side"]]:
has the winning side information for each conversation


The merge connects them by matching conversation_id in advocate_utterances (every utterance knows which conversation it belongs to) with the index of conversations_df (which is the conversation ID). So after the merge, every advocate utterance will have a new column meta.win_side telling us who won that case. Left merge keeps all the advocate utterances even if no match is found.

```python
#create won column
advocate_utterances["won"] = (
    advocate_utterances["meta.side"] == advocate_utterances["meta.win_side"]
).astype(int)

#drop -1 cases
advocate_utterances = advocate_utterances[advocate_utterances["meta.win_side"] != -1]

advocate_utterances["won"].value_counts()
```

*Interpretation*


---



advocate_utterances["meta.side"]:
tells which side the advocate is on (1 for petitioner, 0 for respondent)

advocate_utterances["meta.win_side"]:
tells us which side won the case (1 for petitioner, 0 for respondent)


- So if an advocate is on side 1 and the winning side is 1, they match and the advocate won.

- If an advocate is on side 0 and the winning side is 1, they don't match and the advocate lost.

- The comparison gives us 1's or 0's indicating whether an advocate was on the winning side (1) or on the losing side (0).


```python
#average sentiment scores for winning vs losing advocates
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')

sia = SentimentIntensityAnalyzer()

# get sentiment score for each utterance
sentiments = []
for text in advocate_utterances["text"]:
    #compound = VADER's overall sentiment score for a piece of text, a single number between -1.0 and +1.0
    score = sia.polarity_scores(str(text))["compound"]
    sentiments.append(score)

advocate_utterances.loc[:, "sentiment"] = sentiments

advocate_utterances.groupby("won")["sentiment"].mean()
```

**Findings:**

Sentiment scores are nearly identical between winning and losing advocates which suggests that positivity of language is not a distinguishing factor in Supreme Court oral arguments. Therefore, word choice matters more than tone.

```python
import plotly.express as px

sentiment_means = advocate_utterances.groupby("won")["sentiment"].mean().reset_index()
sentiment_means["outcome"] = ["Losing", "Winning"]

fig = px.bar(sentiment_means, x="outcome", y="sentiment",
             color="outcome",
             color_discrete_map={"Losing": "lightskyblue", "Winning": "#5B2D8E"},
             title="Average Sentiment Score by Outcome",
             labels={"sentiment": "Average Sentiment Score", "outcome": ""})

fig.update_layout(showlegend=False)

#update bar properties to display the sentiment to the 4th decimal
fig.update_traces(texttemplate="%{y:.4f}", textposition="outside")
fig.show()
```

```python
advocate_utterances["year"] = advocate_utterances["meta.case_id"].str.split("_").str[0].astype(int)
advocate_utterances["decade"] = (advocate_utterances["year"] // 10) * 10
advocate_utterances["won_label"] = advocate_utterances["won"].map({0: "Losing", 1: "Winning"})

sentiment_by_decade = advocate_utterances.groupby(["decade", "won_label"])["sentiment"].mean().reset_index()

fig = px.line(sentiment_by_decade, x="decade", y="sentiment",
              color="won_label",
              markers=True,
              color_discrete_map={"Losing": "lightskyblue", "Winning": "#5B2D8E"},
              title="Average Sentiment Score by Decade and Outcome",
              labels={"sentiment": "Average Sentiment Score", "decade": "Decade", "won_label": ""})

fig.update_layout(width=1500)
fig.show()
```

```python
#winning advocate response length
response_lengths = []
for text in advocate_utterances["text"]:
    #split on white space
    length = len(str(text).split())
    response_lengths.append(length)

#get all rows and one column
advocate_utterances.loc[:, "response_length"] = response_lengths

#get justice utterance IDs
justice_ids = set(utterances_df[utterances_df["meta.speaker_type"] == "J"].index)

#filter to advocate utterances that are replying to a justice
responses_to_justices = advocate_utterances[
    advocate_utterances["reply_to"].isin(justice_ids)
]

responses_to_justices.groupby("won")["response_length"].mean()

```

**Findings:**

Winning advocates tend to give longer responses to justice questions on average (65 words) compared to losing advocates (60 words).

```python
response_means = advocate_utterances.groupby("won")["response_length"].mean().reset_index()
response_means["outcome"] = ["Losing", "Winning"]

fig = px.bar(response_means, x="response_length", y="outcome",
             orientation="h",
             color="outcome",
             color_discrete_map={"Losing": "lightskyblue", "Winning": "#5B2D8E"},
             title="Average Response Length by Outcome",
             labels={"response_length": "Average Words per Question", "outcome": ""})

fig.update_traces(texttemplate="%{x:.1f}", textposition="outside")
fig.update_layout(showlegend=False, width=1500)
fig.show()
```

```python
#citing precedent average
citation_counts = []

for text in advocate_utterances["text"]:
    text = str(text).lower()
    count = text.count(" v. ")
    citation_counts.append(count)

advocate_utterances["citation_count"] = citation_counts
advocate_utterances["citation_rate"] = advocate_utterances["citation_count"] / advocate_utterances["response_length"]

advocate_utterances.groupby("won")["citation_rate"].mean()
```

```python
from scipy import stats

winning = advocate_utterances[advocate_utterances["won"] == 1]["citation_rate"]
losing = advocate_utterances[advocate_utterances["won"] == 0]["citation_rate"]

stats.ttest_ind(winning, losing)
```

**Findings:**

The results show winning advocates cite slightly more per word (0.000298) than losing ones (0.000240), though the difference is very small. After doing a t-test to verify that the difference is indeed statistically significant, the difference is too small to conclude much. Lawyers on the winning side do use a bit more precedent, but not much.

```python
citation_means = advocate_utterances.groupby("won")["citation_rate"].mean().reset_index()
citation_means["outcome"] = ["Losing", "Winning"]

fig = px.bar(citation_means, x="outcome", y="citation_rate",
             color="outcome",
             color_discrete_map={"Losing": "lightskyblue", "Winning": "#5B2D8E"},
             title="Average Citation Rate by Outcome",
             labels={"citation_rate": "Average Citation Rate", "outcome": ""})

fig.update_traces(texttemplate="%{y:.6f}", textposition="outside")
fig.update_layout(showlegend=False)
fig.show()
```

```python
#create new dataframe with average citation rate for each decade and outcome combination
citation_by_decade = advocate_utterances.groupby(["decade", "won_label"])["citation_rate"].mean().reset_index()
fig = px.line(citation_by_decade, x="decade",
              y="citation_rate",
              color="won_label",
              markers=True,
              color_discrete_map={"Losing": "lightskyblue", "Winning": "#5B2D8E"},
              title="Average Citation Rate by Decade and Outcome",
              labels={"citation_rate": "Average Citation Rate", "decade": "Decade", "won_label": ""})

fig.update_layout(width=1500)
fig.show()
```

**Analysis:**

Logistic Regression Model to predict whether a lawyer will win or lose based on stretegic features

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

#make one row per advocate per case
features = advocate_utterances.groupby(["speaker", "meta.case_id", "won"]).agg(
    sentiment=("sentiment", "mean"),
    response_length=("response_length", "mean"),
    citation_rate=("citation_rate", "mean"),
).reset_index()

features = features.dropna(subset=["sentiment", "response_length", "citation_rate"])

X_all = features[["sentiment", "response_length", "citation_rate"]]
y_all = features["won"]

scores = cross_val_score(LogisticRegression(class_weight="balanced"),
                         X_all,
                         y_all,
                         cv=5,
                         scoring="accuracy")

#mean accuracy of all features combined
print("Mean accuracy: " + str(scores.mean().round(3)))
```

*Interpretation:*


---

The model correctly predicted whether an advocate won or lost 51.4% of the time. This is only slightly better than randomly guessing (50%), so these features are not the strongest predicters of win rate.

```python
features_list = ["sentiment", "response_length", "citation_rate"]

#loop through each feature
for feature in features_list:
    X_one = features[[feature]]
    y_one = features["won"]

    scores = cross_val_score(LogisticRegression(class_weight="balanced"),
                             X_one,
                             y_one,
                             cv=5,
                             scoring="accuracy")

    print(feature + ":")
    print("  Mean accuracy: " + str(scores.mean().round(3)))
    print("  Std: " + str(scores.std().round(3)))
    print()
```

*Interpretation:*


---

- Sentiment had a 49.7% accuracy rate which is literally worse than random guessing, so there is no predictive power at all.

- Response length had a 51.2% accuracy which is barely above randomly guessing, also suggesting a very weak signal. The high standard deviation on response length (0.07) is also worth noting, because it is very inconsistent across folds, meaning it performs well on some splits and poorly on others, which suggests it's not a reliable predictor.

- Citation rate, on the other hand, had a 61.3% accuracy, which is the highest of the three, meaning that it is the stongest predictor of win rate.


Conclusion:
Advocates who cite more case law are more likely to win.
